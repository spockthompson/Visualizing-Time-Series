## Load in the data 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url ="https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv"
df = pd.read_csv(url)
df
```

## 1) Convert the date column to datetime dtype. 

Double check the result. Is it what you expected? If not, be sure to correct it by entering the appropriate format!

```python
# INCORRECT - does not parse dates correctly
pd.to_datetime(df['date'])
```

```python
# CORRECT - properly recognizes dates and does not interpret them as seconds
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df['date']
```

## 2) Set the date column as the index. 
Inspect the new index of your dataframe. What type of index is it? Does it have a frequency?


```python
df = df.set_index('date')
df
```

## 3) Filter out only the required data:

- Keep only data from the year 2000 or later.
- Keep only the following features:
    - precipitation
    - mean_temp
    - min_temp 
    - max_temp 
    - snow_depth

```python
df = df.loc['2000':, ['precipitation', 'mean_temp', 'min_temp', 
                      'max_temp', 'snow_depth']]

df
```

## 4) Impute any missing values.

Think about each feature and what imputation method makes sense. You may not be able to use the same method for every feature!

```python
df.isna().sum()
```

```python
## interpolating temperature measures
df['mean_temp'] = df['mean_temp'].interpolate()
df['max_temp'] = df['max_temp'].interpolate()
df['min_temp'] = df['min_temp'].interpolate()
```

```python
df.isna().sum()
```

```python
# filling preciptation/snow with 0
df['precipitation'] = df['precipitation'].fillna(0)
df['snow_depth'] = df['snow_depth'].fillna(0)
```

```python
df.isna().sum()
```

```python
from matplotlib import dates as mdates
```

## 5) Plot the the mean, min, and max temps for 2018-2020 as 1 line graph .


- Use a combination of pandas, matplotlib tick formatters, tick locators, and the fig.autofmt_xdate method to:
    - Create a wide figure (e.g. figsize=(12,3))
    - With a title of "Daily Temperatures"
    - With a ylabel of "Degrees"
    - Customize the major xaxis ticks to:
        - Have major xaxis ticks every 1 year.
        - Use 4-digit years for the labels
        - Have the major ticks' labels rotated 90 degrees. 
        
    - Customize the minor xaxis ticks to:
        - Have minor xaxis ticks every 3 months
        - Use 3-letter month names for the labels.

```python
temp_cols = ['mean_temp','min_temp','max_temp']
plot_df = df.loc['2018':,temp_cols]
plot_df
```

```python
## Plot the data
ax = plot_df.plot(figsize=(12,5))
ax.set(title='Daily Temperature', ylabel='Degrees')

# customize major ticks 
major_loc =  mdates.YearLocator(1)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)


# customize minor ticks
minor_loc =  mdates.MonthLocator(interval=3)
minor_fmt = mdates.DateFormatter("%b")

ax.xaxis.set_minor_locator(minor_loc)
ax.xaxis.set_minor_formatter(minor_fmt)

## Final xdate tweaks
fig = ax.get_figure()
fig.autofmt_xdate(which='major',rotation=90)
```

## 6) Plot the precipitation for 2000-2020

- Use a combination of matplotlib tick formatters, tick locators, and the fig.autofmt_xdate method to:
    - Create a wide figure (e.g. figsize=(12,3))
    - With a title of "Daily Precipitation"
    - With a ylabel of "Inches"
    - Customize the major xaxis ticks to:

```python
precip_cols = ['precipitation','snow_depth']
plot_df  = df.loc["2000":,precip_cols]
plot_df
```

```python
ax = plot_df.plot(figsize=(12,3))
ax.set( title='Daily Precipitation & Snow Fall',
      ylabel='Inches')

# customize major ticks 
major_loc =  mdates.YearLocator(5)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)


# customize minor ticks
minor_loc =  mdates.MonthLocator(interval=3)

ax.xaxis.set_minor_locator(minor_loc)


## Final xdate tweaks
fig = ax.get_figure()
fig.autofmt_xdate()
```
